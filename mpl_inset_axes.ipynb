```python
# inset axes with matplotlib
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from numpy import *
from random import random as rand
```

```python
# make some data
xpts = linspace(-10,10,1000)
ypts = array([exp(-x**2)+(.1*rand()-.5) for x in xpts])
```

```python
# the figure and main axes object
fig, ax = plt.subplots()

# the inset axis
ax_in = inset_axes(ax, # parent axis
                   width="40%", # percent of parent width
                   height="40%", # likewise
                   loc=1) # 1 = top right

# plot in both axes
ax.plot(xpts, ypts)
lim1 = 500
lim2 = 900
ax_in.plot(xpts[lim1:lim2],ypts[lim1:lim2])
ax_in.set_ylim((-.6,.6))
ax_in.set_xlim((0,2.5))
plt.show()
```
