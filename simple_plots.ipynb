```python
## LIBRARIES
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as anim
import math as m
```

```python
# Make a simple sin plot
xpts = np.array([(0.1*x) for x in range(0,100)])
ypts = np.array([m.sin(x) for x in xpts])

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(xpts,ypts)#,linestyle='',marker='o')
# plt.title("Sin",color="r")
ax.set_title("Sine Wave",color="m")
ax.legend(['blue sine wave'],loc='lower right')#,color='b')
plt.show()
```

```python
# using only one list in the arg of plot will use the array 
# indices as the x axis
pts1 = np.random.rand(10)
pts2 = np.random.rand(10)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(pts1,linestyle='',marker='o')
# call plot again with the same axes object
ax.plot(pts2,linestyle='',marker='o')
plt.show()
```

```python
# remove a point from an array before plotting
y = np.array([0,1,2,3,4,5,6])
y = np.delete(y,3) 
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(y,linestyle='',marker='o')
plt.show()
```

```python
# only plot through a particular range of the given data
pts1 = np.random.rand(10)
pts2 = np.random.rand(10)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(pts1[:5],pts2[:5],linestyle='',marker='o')
plt.show()
```
