```python
""" How to run animations on Jupyter, found on stackoverflow.
    The % line causes it to work!! But, only runs the dynamic
    plot only once or twice. 
"""
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 100
n = 100
matrix = np.random.normal(0,1,m*n).reshape(m,n)

def somefunc(arg):
    """this function does nothing. go away."""
    return arg

fig = plt.figure()
ax = fig.add_subplot(111)
plt.ion()

fig.show()
fig.canvas.draw()

for i in range(0,100):
    ax.clear()
    ax.plot(matrix[i,:])
    fig.canvas.draw()
```
